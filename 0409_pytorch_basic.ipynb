### matmul (행렬의 곱셈)과 mul (원소별 곱셈)의 차이

```python
t1 = torch.FloatTensor([[1,2],[3,4]])
t2 = torch.FloatTensor([[5],[6]])
print(t1)
print(t2)
print('Shape of Matrix t1: ', t1.shape)
print('Shape of Matrix t2: ', t2.shape)
print(t1.matmul(t2))
print(t1.matmul(t2).shape)
```

```python
t1 = torch.FloatTensor([[1,2],[3,4]])
t2 = torch.FloatTensor([[5],[6]])
print('Shape of Matrix t1: ', t1.shape)
print('Shape of Matrix t2: ', t2.shape)
print(t1)
print(t2)
print(t1*t2)
print(t1.mul(t2))
print(t1.mul(t2).shape)
```

 __'*'연산자를 이용한 broadcasting 가능__

- Max, Argmax

```python
t1 = torch.FloatTensor([1,2,3,4])
print(t1.mean())
```

```python
t1 = torch.FloatTensor([[1,2],[3,4]])
print(t1)
print(t1.mean())
print(t1.mean(dim=0))
print(t1.mean(dim=1))
print(t1.mean(dim=-1))
```

```python
t1 = torch.FloatTensor([[1,2],[3,4]])
print(t1)
print(t1.sum())
print(t1.sum(dim=0))
print(t1.sum(dim=1))
print(t1.sum(dim=-1))
```

```python
t1 = torch.FloatTensor([[1,2,5],[4,3,7]])
print(t1.shape)
print(t1,'\n')
print(t1.max(),'\n')
print(t1.max(dim=0),'\n')
print(t1.max(dim=1),'\n')
print(t1.max(dim=-1))
```

```python
print(t1.max(dim=0).values)
print(t1.max(dim=0).indices) # Argmax
```

- 차원 합치기

```python
x=torch.FloatTensor([[1,2],[3,4]])
y=torch.FloatTensor([[5,6],[7,8]])

print(torch.cat([x,y], dim=0))
print(torch.cat([x,y], dim=0).shape)
print(torch.cat([x,y], dim=1))
print(torch.cat([x,y], dim=1).shape)
```

```python
x=torch.FloatTensor([1,2])
y=torch.FloatTensor([3,4])
z=torch.FloatTensor([5,6])

print(torch.stack([x,y,z]))
print(torch.stack([x,y,z],dim=1))
```
